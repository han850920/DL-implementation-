# Build an MLP and train it with 60000 digits
### here we compare ReLU and swish two different activation func with 60000 digit
### The executing time of "swish" : about 40minute
### The executing time of "ReLU" : about 30 minute
### The final loss of swish is lower than ReLU, but the executing time is longer than ReLU

```python
import numpy as np
import math
%matplotlib inline
import matplotlib.pylab as plt
from model.fully_conn_layer import fcl
from model.conv_layer import cl
from model.MLP import mlp
from model.CNN import cnn

f = open("dataset/mnist_train_60000.csv", 'r')
a = f.readlines()
f.close()

%matplotlib inline
import matplotlib.pylab as plt
x = []
y = []
count=1
for line in a:
    linebits = line.split(',')
    x_line = [int(linebits[i]) for i in range(len(linebits))]
    x.append(x_line[1:])
    y.append(x_line[0])
    
    pass


train_60000_x = np.array(x).T
train_60000_y = np.array(y)
```

# activation_func = "ReLU "

```python
Layers = (784, 50, 10)
BatchSize = 100
EpochCount = 1000
LearningRate = 0.1
Eta = 0.00001

mlp60000_ReLU = mlp(Layers, BatchSize,activation="ReLU")
mlp60000_ReLU.train(train_60000_x, train_60000_y, EpochCount, LearningRate, Eta)
```

```python
plt.plot(mlp60000_ReLU.J)
```

# activation_func =" swish "

```python
Layers = (784, 50, 10)
BatchSize = 100
EpochCount = 1000
LearningRate = 0.1
Eta = 0.00001

mlp60000_swish = mlp(Layers, BatchSize,activation="swish")
mlp60000_swish.train(train_60000_x, train_60000_y, EpochCount, LearningRate, Eta)
```

```python
plt.plot(mlp60000.J)
```

# Build an CNN & MLP and train it with 100 digits
### here we modify parameter and test with 100digit because 60000 is too waste time

```python
# Read in 100 digits

f = open("dataset/mnist_train_100.csv", 'r')
a = f.readlines()
f.close()

# f = figure(figsize=(15,15));
%matplotlib inline
import matplotlib.pylab as plt
x = []
y = []
count=1
for line in a:
    linebits = line.split(',')
    x_line = [int(linebits[i]) for i in range(len(linebits))]
    x.append(x_line[1:])
    y.append(x_line[0])
    imarray = np.asfarray(linebits[1:]).reshape((28,28))
    plt.subplot(10,10,count)
    plt.subplots_adjust(hspace=0.5)
    count += 1
    plt.title("Label is " + linebits[0])
    plt.imshow(imarray, cmap='Greys', interpolation='None')
    pass


x100 = np.array(x).reshape(100, 1, 28, 28) / 256.0 - 0.5
y100 = np.array(y)

test_100_x = x100[:, :, :, :]
test_100_y = y100[:]
val_100_x = x100[:, :, :, :]
val_100_y = y100[:]
```

## BatchSize=25 / EpochCount=10 /LearningRate = 0.1

```python
# Build a CNN identical to LeNet

BatchSize = 25
EpochCount = 10
LearningRate = 0.1
Eta = 0.0001
LeakyRate = 0.01

input_data_spec = [BatchSize, 1, 28, 28]
conv_layer_spec = [{"k_num" : 4, "k_h" : 3, "k_w" : 3, "stride" : 1, "zp" : 0, "maxPoolingHeight" : 2,"maxPoolingWidth" :2},
                   {"k_num" : 16, "k_h" : 3, "k_w" : 3, "stride" : 1, "zp" : 0, "maxPoolingHeight" : 2,"maxPoolingWidth" :2}]
fc_layer_spec = [400, 50, 10]

cnn100 = cnn(input_data_spec=input_data_spec, 
          conv_layer_spec=conv_layer_spec, 
          fc_layer_spec=fc_layer_spec)

cnn100.train(test_100_x, test_100_y, val_100_x, val_100_y, EpochCount, LearningRate, Eta)
```

```python
plt.plot(cnn100.fc_net.J)
plt.plot(cnn100.fc_net.J_val)
```

## BatchSize=25 / EpochCount=10 /LearningRate = 1

```python
# Build a CNN identical to LeNet

BatchSize = 25
EpochCount = 10
LearningRate = 1
Eta = 0.0001
LeakyRate = 0.01

input_data_spec = [BatchSize, 1, 28, 28]
conv_layer_spec = [{"k_num" : 4, "k_h" : 3, "k_w" : 3, "stride" : 1, "zp" : 0, "maxPoolingHeight" : 2,"maxPoolingWidth" :2},
                   {"k_num" : 16, "k_h" : 3, "k_w" : 3, "stride" : 1, "zp" : 0, "maxPoolingHeight" : 2,"maxPoolingWidth" :2}]
fc_layer_spec = [400, 50, 10]

cnn100 = cnn(input_data_spec=input_data_spec, 
          conv_layer_spec=conv_layer_spec, 
          fc_layer_spec=fc_layer_spec)

cnn100.train(test_100_x, test_100_y, val_100_x, val_100_y, EpochCount, LearningRate, Eta)
```

```python
plt.plot(cnn100.fc_net.J)
plt.plot(cnn100.fc_net.J_val)
```
